## 4. Variables and Types
Python supports multiple data types, including:
- `int` - integers (e.g., 2)
- `float` - decimal numbers (e.g., 1.79)
- `str` - strings/text (e.g., 'hello')
- `bool` - booleans (`True` or `False`)

Key concepts:
- Variables make your code reproducible and easier to update
- Python's operators behave differently depending on data type (e.g, `+` adds numbers but concatenates strings)
- You can check a variable's type using the `type()` function

## 5. Variable Assignment 
In Python, a variable lets you store a value using a name.
For example, `x = 5` assigns the value 5 to the variable `x`, allowing you to use `x` in place of 5 in your code.

```python
# Create a variable savings
savings = 100

# Print out savings
print(savings)
```

## 6. Calculations with Variables - Exercise
You've now created a savings variable, so let's start saving!

Instead of calculating with the actual values, you can use variables instead.

How much money would you have saved four months from now, if you saved $10 each month?

```python
# Create the variables monthly_savings and num_months
monthly_savings = 10
num_months = 4


# Multiply monthly_savings and num_months
new_savings = monthly_savings * num_months

# Print new_savings
print(new_savings)
```

## 8. Operations with other types
In Python, variables have different types, which you can check using `type()`. Each type behaves differently - for example, adding strings combins them, while adding integers or booleans performs arithmetic.

```python
savings = 100
new_savings = 40

# Calculate total_savings using savings and new_savings
total_savings = savings + new_savings
print(total_savings)

# Print the type of total_savings
print(type(total_savings))
```
